```python
import qiskit as qk
import math
```

```python
def GHZcirc(n,basis,cname='circuit0'):
    
    qr = qk.QuantumRegister(n)
    cr = qk.ClassicalRegister(n)
    ghz = qk.QuantumCircuit(qr,cr,name=cname)
    
    ghz.h(qr[0])
    for i in range(n-1):
        ghz.cx(qr[i],qr[i+1])
    
    if basis == 'X':
        for j in range(n):
            ghz.h(qr[j])
    elif basis == 'Y':
        for j in range(n):
            ghz.u3(math.pi/2,-math.pi/2,math.pi/2,qr[j])
        
    return ghz
```

```python
n = 7
basis = 'Y'
circ = GHZcirc(n,basis)
circ.draw()
```
